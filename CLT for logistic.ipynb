```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_size = [1, 5, 20, 50, 100, 200, 400]
loc, scale = 10, 1
num_samples = 20000

def normal_dist(mean, sd, x):
    return (1/(sd*np.sqrt(2*np.pi)))*np.exp(-0.5*((x-mean)/sd)**2)
```

```python
for ss in sample_size:
    sample_means = []
    plt.clf()

    for _ in range(num_samples):
        distribution = np.random.logistic(loc, scale, ss)
        sample_means.append(np.mean(distribution))

    meanofmeans = np.mean(sample_means)
    sdofmeans = np.std(sample_means)
    x = np.linspace(np.min(sample_means), np.max(sample_means), 100)
    normal = normal_dist(meanofmeans, sdofmeans, x)

    plt.hist(sample_means, bins=50, density=True, color='b', alpha = 0.7, label='Sample Means')
    plt.plot(x, normal, c = "black", lw = 2, label = "Normal Distribution")
    plt.grid()
    plt.title(f'Sample Size = {ss}, Logistic Distribution')
    plt.xlabel('Sample Means')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.savefig(f"logistic-{ss}.png", dpi = 100)
```
