We see the only quantities showing any significant correlation with measured solubility are <br>
1. Predicted solubility <br>
2. Molecular weight

Lets start by looking at predicted solubility first. We will fit a best fit line through the data points to study the linear dependence.

```python
x = df['measured log solubility in mols per litre']
y = df['ESOL predicted log solubility in mols per litre']

popt, pconv = np.polyfit(x, y, 1, cov=True)
xvals = np.linspace(np.min(x)-1, np.max(x)+1)
yvals = [popt[0]*x + popt[1] for x in xvals]
```

We get the best fit parameters as follows : <br>
<br>
a = 0.728 $\pm$ 0.010 <br>
b = -0.766 $\pm$ 0.037 <br>
<br>
We see our fitted parameters a and b have confidence of $1.37\%$ and $4.83\%$ so the fit is not bad.

```python
np.diag(np.sqrt(pconv)), popt[0], popt[1]
```

Plotting the results

```python
plt.scatter(x, y, c = "black",s = 5)
plt.plot(xvals, yvals, c = "red", lw = 2)
plt.xlabel("Actual Solubility")
plt.ylabel("Predicted Solubility")
plt.legend(["Data Points", "Best-Fit Line"])
plt.text(-12.2, -10.2, f"y = {np.round(popt[0], 3)}x{np.round(popt[1], 3)}", c = "red")
```

The least square error comes out to be 0.828
